## Forge datasets

This data set symbolize 2-class classification dataset

```python
import mglearn
```

```python
# generates the data set
X, y = mglearn.datasets.make_forge()
```

```python
# plot the datasets
import matplotlib.pyplot as plt
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["class 0", "class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

Just as above shown, by default, KNN will take only 1 neighbors to determine which class it is belongs to. If we changed the parameters, the prediction also may change. 
Look at the 1st input point, with 1 neighbor it is part of `training class 0`, with 3 neighbors, it is `training class 1`

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

Then this prompt a question, how many neighbors do we consider?? Are we underfitting or overfitting our prediction?

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
fig, axes = plt.subplots(1, 4,figsize=(10,3))
for n_neighbor, ax in zip([1,2,4,6], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbor).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1],y, ax=ax)
    ax.set_title("{} neighbors".format(n_neighbor))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
axes[0].legend(["class 0", "class 1"],loc=3)
```

The more neighbors that we use, the smoother the curve. In another word, the less neighbors we consider, the more complex model will be.
